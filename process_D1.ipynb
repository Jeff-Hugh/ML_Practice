```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
```

```python
# 读取excel数据，第一行为列名
file = "data/data.xlsx"
data = pd.read_excel(file, header=0)

# 截取数据的第7列至“总通最大值”
data = data.iloc[:, 6:-3]

# 打印数据的前几行
print(data.head()) 
```

```python
# 检查并处理缺失值（NaNs）和异常值（outliers）
def preprocess_data(df):
    # 检查缺失值
    if df.isnull().values.any():
        # 如果存在缺失值，打印缺失值的索引
        print("缺失值的索引：")
        print(df[df.isnull().any(axis=1)].index.tolist())

        print("数据中存在缺失值，正在处理...")

        df = df.fillna(df.mean())  # 用均值填充缺失值
        print("缺失值已处理。")
    else:
        print("数据中没有缺失值。")

    # 检查异常值
    for column in df.select_dtypes(include=[np.number]).columns:
        if (df[column] < 0).any():
            print(f"列 {column} 存在异常值（负数），正在处理...")
            df[column] = df[column].clip(lower=0)

    print("异常值已处理。")
    return df

data = preprocess_data(data)
```

```python
# 进行Z-score标准化
from sklearn.preprocessing import StandardScaler
print("原始数据的统计信息:")
print(data.describe())

# 2. 创建并使用 StandardScaler
# 创建StandardScaler的实例
scaler = StandardScaler()

# 使用 .fit_transform() 一步完成拟合和转换
# 注意：StandardScaler的输出是一个NumPy数组，而不是Pandas DataFrame
scaled_data_numpy = scaler.fit_transform(data)
# 为了方便查看，可以将其转换回DataFrame
scaled_data = pd.DataFrame(scaled_data_numpy, columns=data.columns)

print("\n-------------------------------------------------")
print("\n标准化后数据的统计信息:")
print(scaled_data.describe())
```

```python
# 1. 计算Spearman相关系数矩阵
correlation_matrix = scaled_data.corr(method='spearman')

# 2. 使用热力图进行可视化
plt.rcParams['font.sans-serif'] = ['SimHei'] # For Windows
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12, 10))
seaborn.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Spearman Correlation Heatmap')
plt.show()
```

```python
# 选择所有指标的平均值进行分析，即第1、7、13、19、25、31、36列
selected_data = scaled_data.iloc[:, [0, 6, 12, 18, 24, 35]]
print("选择后的数据形状:", selected_data.shape)
```

```python
# 1. 计算Spearman相关系数矩阵
correlation_matrix = selected_data.corr(method='spearman')

# 2. 使用热力图进行可视化
plt.rcParams['font.sans-serif'] = ['SimHei'] # For Windows
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(12, 10))
seaborn.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Spearman Correlation Heatmap')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x, y = selected_data.iloc[:, :-1], selected_data.iloc[:, -1]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 多元线性回归模型
model = LinearRegression()

# 在训练数据上训练模型
model.fit(x_train, y_train)
print("模型训练完成。")
```

```python
# 获取模型学到的系数 (beta_1, beta_2, ...)
coefficients = model.coef_
# 获取模型学到的截距 (beta_0)
intercept = model.intercept_

print(f"学到的系数 (β1, β2... β5): {coefficients}")
print(f"学到的截距 (β0): {intercept}")

# 在测试集上进行预测
y_pred = model.predict(x_test)

# 评估模型性能
# 均方误差 (Mean Squared Error, MSE) - 越小越好
mse = mean_squared_error(y_test, y_pred)
# R^2 决定系数 (R-squared) - 越接近1越好
r2 = r2_score(y_test, y_pred)

print(f"\n模型在测试集上的性能评估:")
print(f"均方误差 (MSE): {mse:.2f}")
print(f"R² 分数: {r2:.2f}")
```

```python
from sklearn.model_selection import validation_curve
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# 我们使用一个 Pipeline，它将多项式特征转换和线性回归打包在一起
# 这使得我们可以方便地将 'degree' 作为超参数进行调整
def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                         LinearRegression(**kwargs))

def rms_error(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

rms_scorer = make_scorer(rms_error, greater_is_better=False)


# 3. 设置要测试的超参数范围
# 我们要测试多项式的阶数 (degree)
# 注意：在 Pipeline 中，超参数的名称是 '组件名__参数名'
degree = 3
param_name = 'polynomialfeatures__degree'
param_range = np.arange(1, degree + 1)

# 4. 使用 validation_curve 函数
# scoring='neg_mean_squared_error': 使用负均方误差。scikit-learn约定分数越高越好，
# 所以它取MSE的负值。我们后面绘图时可以再把它变回正的。
# cv=5: 使用5折交叉验证
train_scores, test_scores = validation_curve(
    PolynomialRegression(), x, y,
    param_name=param_name,
    param_range=param_range,
    cv=5,
    scoring=rms_scorer
)

# 5. 计算平均值和标准差，并绘图
# 分数取反，变回正的MSE
train_scores_mean = -np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = -np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

# 绘制验证曲线
plt.figure(figsize=(10, 6))
plt.title("Validation Curve for Polynomial Regression")
plt.xlabel("Degree of Polynomial")
plt.ylabel("Mean Squared Error (MSE)")
plt.grid(True)

# 绘制训练得分曲线
plt.plot(param_range, train_scores_mean, label="Training score", color="darkorange", lw=2)
# 填充标准差区域
plt.fill_between(param_range, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.2,
                 color="darkorange")

# 绘制交叉验证得分曲线
plt.plot(param_range, test_scores_mean, label="Cross-validation score", color="navy", lw=2)
# 填充标准差区域
plt.fill_between(param_range, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.2,
                 color="navy")

plt.legend(loc="best")
plt.show()
```
